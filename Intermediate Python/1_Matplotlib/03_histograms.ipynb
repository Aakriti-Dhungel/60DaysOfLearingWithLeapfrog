## Customization
### choice depends on - data , story you want to tell

```python
import matplotlib.pyplot as plt

values = [0, 0.6, 1.4, 1.6, 2.2, 2.5, 2.6, 3.2, 3.5, 3.9, 4.2, 6]

plt.hist(values, bins=3, 
         color='skyblue',       # Bar color
         edgecolor='black',     # Bar border color
         alpha=0.7)             # Transparency

plt.title('Histogram of Values', fontsize=14, fontweight='bold')
plt.xlabel('Value Range', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)  # Horizontal gridlines for y-axis

plt.show()
```
