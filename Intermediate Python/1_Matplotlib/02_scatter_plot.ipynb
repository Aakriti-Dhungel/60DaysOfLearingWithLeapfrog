```python
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2010]
population = [2.519,3.692,5.263,6.972] # billion

year = [1800, 1825, 1850, 1875, 1900, 1925, 1960, 1980, 2000, 2020] + year
population = [1.0, 1.1, 1.262, 1.4, 1.65, 2.0, 3.0, 4.0, 6.0, 7.8] + population

plt.scatter(year, population, marker='o', c='blue', alpha=0.8, s=[p*100 for p in population])
plt.xlabel("Year")
plt.ylabel("Population (in billions)")
plt.title("World Population Over Time")
plt.grid(True)
plt.xticks(rotation=45)
plt.yticks([0,2,4,6,8,10], ['0B', '2B', '4B', '6B', '8B', '10B'])

plt.text(1950, 2.5, 'Mid 20th Century') # plt.text(x,y,"text") - calls plt.grid
plt.text(2020, 7.8, 'Recent Year')

plt.show()
```
